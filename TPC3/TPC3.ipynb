```python
def criarListaIn():
    lista=[]
    n=int(input("Introduza o número de elementos da lista:"))
    i=1
    while len(lista)<n:
        elem=input("Introduza o elemento "+str(i)+":")
        lista.append(elem)
        i=i+1
    return lista

def criarListaA():
    from random import randrange
    lista=[]
    n=int(input("Introduza o número de elementos da lista:"))
    i=1
    while len(lista)<n:
        elem=randrange(0,101)
        lista.append(elem)
        i=i+1
    return lista

def maiorElemento(lista):
    maior=lista[0]
    for i in range(len(lista)):
        if lista[i]>maior:
            maior=lista[i]
    return maior

def bubbleSort(lista):
    ordem=False
    while not ordem:
        ordem=True
        for i in range (len(lista)-1):
            if int(lista[i]) > int(lista[i+1]):
                ordem=False
                lista[i], lista[i+1]=lista[i+1], lista[i]
    return(lista)  

def menu():
    print("""
    (1)Gerar uma lista manualmente
    (2)Gerar uma lista de valores aleatórios
    (3)Apresentar o maior elemento da lista
    (4)Ordenar a lista
    (5)Escrever a lista corrente
    (6)Limpar a lista
    (0)Sair""")
opcao = '-1'
while opcao != "0":
    menu()
    opcao= input("Introduza uma opção:")
    if opcao=="1":
        lista = criarListaIn()
        print (lista)
    elif opcao=="2":
        lista=criarListaA()
        print(lista)
    elif opcao=="3":
        print(maiorElemento(lista))
    elif opcao=="4":
        print(bubbleSort(lista))
    elif opcao=="5":
        print(lista)
    elif opcao=="6":
        lista.clear()
    elif opcao=="0":
        print ("Até à próxima!!")
    else:
        print("Opção inválida :(")
```

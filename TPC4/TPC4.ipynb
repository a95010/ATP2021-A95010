```python
def  criarFrac(numerador,denominador):
    return (numerador,denominador)

def verFrac(f):
    n, d = f
    print(str(n) , "/" , (d))

def mdc(a,b):
    if a<b:
        return mdc(b,a)
    elif a%b==0:
        return b
    else:
        return mdc(a,a%b)
    
def simplificarFrac(f):
    n,d =f
    m= mdc (n,d)
    return (n/m,d/m)

def somaFrac(f1, f2):
    n1,d1 = f1
    n2,d2 = f2
    return (simplificarFrac(n1*d2 + n2*d1, d1*d2))

def subtraiFrac(f1, f2):
    n1,d1 = f1
    n2,d2 = f2
    return (simplificarFrac(n1*d2 - n2*d1, d1*d2))

def multiplicarFrac(f1, f2):
    n1,d1 = f1
    n2,d2 = f2
    return (simplificarFrac(n1*n2, d1*d2))

def dividirFrac(f1, f2):
    n1,d1 = f1
    n2,d2 = f2
    return (simplificarFrac(n1*d2 + n2*d1, d1*d2))

```
